##5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate the Synthetic Dataset
np.random.seed(42)  # For reproducibility

# Create some correlated data
size = 100  # Number of samples

# Generate random data for feature A
A = np.random.rand(size)

# Create B, which is correlated with A
B = A + np.random.normal(0, 0.1, size)

# Create C, which is negatively correlated with A and B
C = -A + np.random.normal(0, 0.1, size)

# Create D, which is a random feature
D = np.random.rand(size)

# Step 2: Create a DataFrame
df = pd.DataFrame({
    'A': A,
    'B': B,
    'C': C,
    'D': D
})

# Step 3: Calculate the Correlation Matrix
corr_matrix = df.corr()

# Step 4: Visualize the Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

##3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

heatmap_data = df.pivot_table(index = 'Day', columns = 'Month', values = 'Sales', aggfunc = 'sum').fillna(0)

fig = px.imshow(heatmap_data,
                labels={'x':'Month',
                       'y':'Day',
                       'color' : 'Sales'},
                x = heatmap_data.columns,
                y = heatmap_data.index)

fig.show()
```

##4. Using the given x and y data, generate a 3D surface plot to visualize the function

```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly.io as pio

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid
x, y = np.meshgrid(x, y)

# Calculate z using the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays to create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot
surface = go.Surface(x=x, y=y, z=z)
layout = go.Layout(
    title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)
fig = go.Figure(data=[surface], layout=layout)

# Show the plot
pio.show(fig)
```

#BOKEH

##1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Display plots in the notebook
output_notebook()

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

##2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource, ColorBar
from bokeh.layouts import layout

# Display plots in the notebook
output_notebook()

# Generate random data
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points) * 100
y = np.random.rand(n_points) * 100
sizes = np.random.rand(n_points) * 50 + 10
colors = np.random.rand(n_points) * 100

df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

source = ColumnDataSource(df)

color_mapper = linear_cmap(field_name='colors', palette='Viridis256', low=min(colors), high=max(colors))

p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='x', y_axis_label='y')

p.scatter('x', 'y', size='sizes', color=color_mapper, alpha=0.6, source=source, legend_label='Random Points')

color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

show(p)
```

##3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Display plots in the notebook
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Convert the data into a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Define a color map
palette = ['#f46d43', '#fdae61', '#fee08b', '#d73027']
color_map = factor_cmap(field_name='fruits', palette=palette, factors=fruits)

# Create a new plot with categorical x-axis
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add bars to the plot
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=color_map)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
##
```

##4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Display plots in the notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency', height=350, width=600)

# Add a quad glyph to represent the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

##5. Create a Bokeh heatmap using the provided dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

p = figure(title="Simple Heatmap", x_axis_label='X', y_axis_label='Y')

p.rect(x_flat, y_flat, width=0.1, height=0.1, color="navy", alpha=z_flat)

show(p)
```
